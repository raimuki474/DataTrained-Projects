```python
## Importing required Libraries for working on the dataset

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
world_happiness_data = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/DSData/master/happiness_score_dataset.csv')
world_happiness_data
```

```python
world_happiness_data.info()
```

```python
world_happiness_data.describe()
```

```python
world_happiness_data.columns
```

```python
world_happiness_data1 = world_happiness_data[['Happiness Score', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)','Generosity']]
world_happiness_data1.head()
```

```python
plt.figure(figsize = (8, 8))
sns.heatmap(world_happiness_data1.corr(), annot = True, cmap = 'coolwarm')
```

```python
variable_pairplot = world_happiness_data1.dropna()
sns.pairplot(variable_pairplot, height = 2)
```

<p><strong><h4><u>OBSERVATIONS</u> :</h4></strong></p>
<ul>
    <li> From the above heatmap,following factors contribute to happiness score: </li>
    <ol><li>Economy (GDP per Capita)</li>
        <li>Family</li>
        <li>Health (Life Expectancy) </li> </ol>
    <li>Increase in Economy (GDP per Capita) has strong positive effect on Health (Life Expentancy) and vice-versa. </li>
    <li> There is no significance correlation between Economy and Generosity </li>
    <li> There is no strong negative correlation between any of the parameters </li>
    </ul>
