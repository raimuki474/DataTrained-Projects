```python
# Importing necessary libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('C:/Users/mukesh.rai/DataTrained/Evaluation Projects/avocado.csv')
dataset
```

```python
# We need to identify if there is any null variable in the given dataset

dataset.isnull().sum()
```

```python
# understanding the type of variable in the given dataset
dataset.info()
```

From above details, we can observe that - out of the total 16468 entries, there are only 1517 non-null records.
Hence we have remove all the remaining null records from the dataset

```python
data = dataset.copy()      # Creating a copy of original dataset
data = data.dropna()       # Removing all the observations having values as NaN
```

```python
data.isnull().sum()
```

Now, as we can observe that, all null values has been removed from the dataset. We need to check on the datatype for each of the variable, since, we have already seen that "Date" variable has float datatype instead of int. Hence, we need to re-check and accordingly adjust the applicable datatype

```python
data.info()
```

```python
data = data.astype({"year" : int})
data.info()
```

We need to drop the "Unnamed : 0" as it doesn't have any important data.

```python
data = data.drop(["Unnamed: 0"], axis = True)   # Removing the Unnamed column
data
```

By visiting the website of avocado dataset, we can rename the numeric title variables as mentioned below :
- 4046 = Small Hass
- 4225 = Large Hass
- 4770 = Xlarge Hass

```python
data = data.rename(index = str, columns ={"4046" : "Small Hass", "4225" : "Large Hass", "4770" : "XLarge Hass" })
data
```

```python
# Understanding the mean of 'averageprice' variable for different years

data.groupby('year')['AveragePrice'].mean()
```

```python
# On the other hand, we need to find number of avocado sold in these 2 years

data.groupby('year')['type'].value_counts()
```

```python
data['month'] = pd.DatetimeIndex(data['Date']).month
data
```

```python
# Visualizing the change in averageprice over time
data.sort_values(by = ['Date'], inplace = True, ascending = True)
plt.figure(figsize = (20, 10))
sns.lineplot(x = 'month', y = 'AveragePrice', hue = 'year', data = data)
plt.show()
```

From the above graph, we can infer that :
- The AveragePrice has been nearly same for the first 5 months in both the years
- However, in the month of June - 2016, the price exponentially increased and remained on the higher side for the rest of the months in the year 2016, except for Aug - 2015 and 2016, where price has been same for both years
- There is high variation in AveragePrice in 2016, while it tends to approx. same in 2015 

```python
# Let us understand the dependence between the various variables using correlation 
plt.figure(figsize = (15, 8))
sns.heatmap(data.drop(['Date', 'month', 'region', 'type', 'year'], axis = 1).corr(), cmap = 'coolwarm', annot = True)
```

From the above heatmap, there is high collinearity between the independent variable. However, it does have any effect on dependent variable i.e. averageprice

```python
data_v1 = data.drop(['Date', 'region', 'type'], axis = 1)
data_v1
```

```python
# Using Linear regression model
X = data_v1.drop(['AveragePrice'], axis = 1)
y = data_v1['AveragePrice']
X
```

```python
# Splitting data in training and testing part
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
predict_lr = lr.predict(X_test)
```

```python
from sklearn import metrics
print('MAE : ', metrics.mean_absolute_error(y_test, predict_lr))
print('MSE : ', metrics.mean_squared_error(y_test, predict_lr))
print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_test, predict_lr)))
```

```python
plt.scatter(x = y_test, y = predict_lr)
```

```python
# Let us find the various parameters of Linear Regression model for given dataset
import statsmodels.api as sm
X_data = sm.add_constant(X)
lr_model = sm.OLS(y, X_data).fit()
print(lr_model.summary())
```

### KNN Regressor

```python
# Using KNN Regressor Model for predicting the AveragePrice of Avocado

from sklearn import neighbors
from math import sqrt

knn = neighbors.KNeighborsRegressor()
knn.fit(X_train, y_train)
pred_knn = knn.predict(X_test)
```

```python
# Calculating RMSE value of KNN Model

error = sqrt(metrics.mean_squared_error(y_test, pred_knn))
print('RMSE value of KNN Model is : ', error)
```

```python
# Finding the R2 value of KNN Model

knn.score(X_train, y_train) 
```

```python
plt.scatter(x = y_test, y = pred_knn)
plt.xlabel("Y Test")
plt.ylabel("Predicted Y")
```

### DecisionTree Regressor Model :

```python
from sklearn.tree import DecisionTreeRegressor

decision_tr = DecisionTreeRegressor()
decision_tr.fit(X_train, y_train)
pred_dtr = decision_tr.predict(X_test)
```

```python
print('MAE : ', metrics.mean_absolute_error(y_test, pred_dtr))
print('MSE : ', metrics.mean_squared_error(y_test, pred_dtr))
print('RMSE : ', sqrt(metrics.mean_squared_error(y_test, pred_dtr)))
```

```python
# Finding the R2 value of DecisionTree Regressor Model

print(decision_tr.score(X_train,y_train))
```

```python
plt.scatter(x = y_test, y = pred_dtr)
plt.xlabel("Y Test")
plt.ylabel("Predicted Y")
```

### RandomForestRegressor :

```python
from sklearn.ensemble import RandomForestRegressor

random_fr = RandomForestRegressor()
random_fr.fit(X_train, y_train)
pred_rfr = random_fr.predict(X_test)
```

```python
print('MAE : ', metrics.mean_absolute_error(y_test, pred_rfr))
print('MSE : ', metrics.mean_squared_error(y_test, pred_rfr))
print('RMSE : ', sqrt(metrics.mean_squared_error(y_test, pred_rfr)))
```

```python
# Finding the R2 value of RandomForest Regressor Model

print(random_fr.score(X_train,y_train))
```

```python
plt.scatter(x = y_test, y = pred_rfr)
plt.xlabel("Y Test")
plt.ylabel("Predicted Y")
```

```python
sns.distplot((y_test, pred_rfr), bins = 15)
```

Therefore, from all the available model, we can infer that RandomForest Regressor Model is best fitted model as it's RMSE is the lowest. Hence, we can predict the averageprice of avocado using RandomForest Regressor technique
