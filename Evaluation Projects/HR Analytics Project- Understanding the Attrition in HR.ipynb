```python
# understanding the relationship between the attrition and salary

attrition_yes = dataset_v1[dataset_v1['Attrition'] == 'Yes']
attrition_no = dataset_v1[dataset_v1['Attrition'] == 'No']

income_attr_y = attrition_yes.groupby(['Department'])['MonthlyIncome'].mean()
income_attr_n = attrition_no.groupby(['Department'])['MonthlyIncome'].mean()
```

```python
ig, axes = plt.subplots(1,2, figsize=(20, 6))

axes[0].bar(income_attr_y.index, income_attr_y, color='brown')
axes[0].set_title('Attrited Employees')

axes[1].bar(income_attr_n.index, income_attr_n, color='green')
axes[1].set_title('Not Attrited Employees')
```

from the above bar plot, we can infer that - Attrition is more when there is disparity between the employees within the same department. Hence, employee with lower salary has higher probability of leaving the firm and on the other hand, employee with higher salary tends to stay

```python
# Using one hot encoding
dataset_v2 = pd.get_dummies(dataset_v1)
dataset_v2.columns
```

```python
X = dataset_v2.drop(['Attrition_No', 'Attrition_Yes', 'Attrition'], axis = 1)
dataset_v2['Attrition'] = dataset_v2['Attrition_Yes']
y = dataset_v2['Attrition']
```

```python
# Splitting data in training and testing part
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train1 = pd.DataFrame(sc.fit_transform(X_train))
X_test1 = pd.DataFrame(sc.transform(X_test))
X_train1.columns = X_train.columns.values
X_test1.columns = X_test.columns.values
X_train1.index = X_train.index.values
X_test1.index = X_test.index.values
X_train = X_train1
X_test = X_test1
```

### DecisionTree Technique :

```python
# Using Decision Tree technique

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
dt_classifier = DecisionTreeClassifier(criterion="gini", max_depth=5, min_samples_split = 2,  min_samples_leaf=1, random_state=42) 
dt_classifier.fit(X_train, y_train)
```

```python
predict_y = dt_classifier.predict(X_test)

print('Accuracy Score using DecisionTree Classifier is : ', accuracy_score(y_test, predict_y))
print('Precision Score using DecisionTree Classifier is : ', precision_score(y_test, predict_y))
print('Recall Score using DecisionTree Classifier is : ', recall_score(y_test, predict_y))
print('F1 Score using DecisionTree Classifier is : ', f1_score(y_test, predict_y))
```

### LogisticRegression Classifier :

```python
# Using Logistic Regression Technique

from sklearn.linear_model import LogisticRegression

logistic_reg_clf = LogisticRegression(solver='liblinear', penalty='l1')
logistic_reg_clf.fit(X_train, y_train)
```

```python
predict_lr = logistic_reg_clf.predict(X_test)

print('Accuracy Score using LogisticRegression Classifier is : ', accuracy_score(y_test, predict_lr))
print('Precision Score using LogisticRegression Classifier is : ', precision_score(y_test, predict_lr))
print('Recall Score using LogisticRegression Classifier is : ', recall_score(y_test, predict_lr))
print('F1 Score using LogisticRegression Classifier is : ', f1_score(y_test, predict_lr))
```

### Randomforest Classifier :

```python
from sklearn.ensemble import RandomForestClassifier

randomforest_class = RandomForestClassifier(n_estimators = 100, bootstrap=False)
randomforest_class.fit(X_train, y_train)
```

```python
predict_rfc = randomforest_class.predict(X_test)

print('Accuracy Score using RandomForest Classifier is : ', accuracy_score(y_test, predict_rfc))
print('Precision Score using RandomForest Classifier is : ', precision_score(y_test, predict_rfc))
print('Recall Score using RandomForest Classifier is : ', recall_score(y_test, predict_rfc))
print('F1 Score using RandomForest Classifier is : ', f1_score(y_test, predict_rfc))
```

By comparing above models, Logistic Regression has better edge than RandomForest
