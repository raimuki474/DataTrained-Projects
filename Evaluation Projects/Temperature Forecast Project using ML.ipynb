```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
dataset_v1 = pd.read_csv('temperature.csv')
dataset_v1
```

```python
dataset_v1.info()
```

```python
dataset_v1.isnull().sum()
```

```python
# Converting the Date format into Year, Month and Date Values

dataset_v2 = dataset_v1.copy()

dataset_v2 = dataset_v2.dropna(subset = ['Date'])

dataset_v2['Year'] = pd.DatetimeIndex(dataset_v2['Date']).year.astype(int)
dataset_v2['Month'] = pd.DatetimeIndex(dataset_v2['Date']).month
dataset_v2['Day'] = pd.DatetimeIndex(dataset_v2['Date']).day
```

```python
dataset_v2 = dataset_v2.drop(columns = ['Date', 'station'])
dataset_v2.head()
```

```python
dataset_v2 = dataset_v2.apply(lambda x: x.fillna(x.mean()),axis=0)
```

```python
dataset_v2.isnull().sum()
```

```python
plt.figure(figsize=(16,8))

plt.subplot(2, 2, (1,2))
sns.distplot(dataset_v2['Present_Tmax'], kde=True, color='y')

plt.subplot(2, 2, (3,4))
sns.distplot(dataset_v2['Present_Tmin'], kde=True, color='g')

print("From the given dataset :")
print('Minimum Tmax value is {} and Maximum Tmax value is {}'.format(dataset_v2['Present_Tmax'].min(),dataset_v2['Present_Tmax'].max()))
print('Minimum Tmin value is {} and Maximum Tmin value is {}'.format(dataset_v2['Present_Tmin'].min(),dataset_v2['Present_Tmin'].max()))
```

Both the features as per the above graph seems to be normally distributed. Also, we have obtained min and max values of both the features

```python
plt.figure(figsize=(16,8))

plt.subplot(2, 2, (1,2))
sns.distplot(dataset_v2['LDAPS_RHmax'], kde=True, color='y')

plt.subplot(2, 2, (3,4))
sns.distplot(dataset_v2['LDAPS_RHmin'], kde=True, color='g')

print("From the given dataset :")
print('Minimum Tmax value is {} and Maximum Tmax value is {}'.format(dataset_v2['LDAPS_RHmax'].min(),dataset_v2['LDAPS_RHmax'].max()))
print('Minimum Tmin value is {} and Maximum Tmin value is {}'.format(dataset_v2['LDAPS_RHmin'].min(),dataset_v2['LDAPS_RHmin'].max()))
```

LDAPS_RHmax seems to be left skewed while LDAPS_RHmin is right skewed

```python
plt.figure(figsize=(16,8))

plt.subplot(2, 2, (1,2))
sns.distplot(dataset_v2['Next_Tmax'], kde=True, color='y')

plt.subplot(2, 2, (3,4))
sns.distplot(dataset_v2['Next_Tmin'], kde=True, color='g')

print("From the given dataset :")
print('Minimum Tmax value is {} and Maximum Tmax value is {}'.format(dataset_v2['Next_Tmax'].min(),dataset_v2['Next_Tmax'].max()))
print('Minimum Tmin value is {} and Maximum Tmin value is {}'.format(dataset_v2['Next_Tmin'].min(),dataset_v2['Next_Tmin'].max()))
```

Both the distribution are nearly normally distributed

```python
plt.figure(figsize=(24,15))
sns.heatmap(dataset_v2.corr(), annot=True, cmap='coolwarm')
```

- There is high correlation between "Next_Tmin" and "LDAPS_Tmin_lapse"
- Another strong correlation is observed between "Next_Tmax" and "LDAPS_Tmax_lapse"
- It is followed by correlation between "Next_Tmin" and "Present_Tmin"

```python
## Preparing the features and target variable

X = dataset_v2.drop(['Next_Tmax', 'Next_Tmin'], axis=1)

y_Tmax = dataset_v2['Next_Tmax']
y_Tmin = dataset_v2['Next_Tmin']
```

```python
X.info()
```

```python
# Splitting the dataset into training and test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_Tmax,test_size = 0.3, random_state=0)
```

```python
# Using Linear Regression

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
predict_lr = lr.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
kfold = StratifiedKFold(n_splits=20)

print('MAE : ', metrics.mean_absolute_error(y_test, predict_lr))
print('MSE : ', metrics.mean_squared_error(y_test, predict_lr))
print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_test, predict_lr)))
print('R2 Score : ', metrics.r2_score(y_test, predict_lr))
```

```python
# Plotting the output graph for Linear Regression Model

sns.distplot(y_test-predict_lr)
plt.show()

plt.scatter(y_test, predict_lr, alpha = 0.5)
plt.xlabel("Test (y)")
plt.ylabel("Prediction (y)")
plt.show()
```

```python
#Using Decision Tree Model

from sklearn.tree import DecisionTreeRegressor

decision_tr = DecisionTreeRegressor(random_state = 0)
decision_tr.fit(X_train, y_train)
pred_dtr = decision_tr.predict(X_test)
```

```python
print('MAE : ', metrics.mean_absolute_error(y_test, pred_dtr))
print('MSE : ', metrics.mean_squared_error(y_test, pred_dtr))
print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_test, pred_dtr)))
print('R2 Score : ', metrics.r2_score(y_test, pred_dtr))
```

```python
# Plotting the output graph for Decision Tree Regression Model

sns.distplot(y_test-pred_dtr)
plt.show()

plt.scatter(y_test, pred_dtr, alpha = 0.5, color = "DarkBlue")
plt.xlabel("Test (y)")
plt.ylabel("Prediction (y)")
plt.show()
```

```python
# Random Forest Regression

from sklearn.ensemble import RandomForestRegressor

random_fr = RandomForestRegressor(n_estimators = 400, min_samples_split = 15, min_samples_leaf = 2, max_features = 'auto', max_depth = 30)
random_fr.fit(X_train, y_train)
pred_rfr = random_fr.predict(X_test)
```

```python
print('MAE : ', metrics.mean_absolute_error(y_test, pred_rfr))
print('MSE : ', metrics.mean_squared_error(y_test, pred_rfr))
print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_test, pred_rfr)))
print('R2 Score : ', metrics.r2_score(y_test, pred_rfr))
```

```python
# Plotting the output graph for Random Forest Regression Model

sns.distplot(y_test-pred_rfr)
plt.show()

plt.scatter(y_test, pred_rfr, alpha = 0.5, color = "DarkBlue")
plt.xlabel("Test (y)")
plt.ylabel("Prediction (y)")
plt.show()
```

Hence, from the above models - Random Forest is the best one with R2 score of 0.894 for predicting "Tmax" value
